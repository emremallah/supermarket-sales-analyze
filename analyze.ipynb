```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


city_counts = df['City'].value_counts()

plt.figure(figsize=(10,5))
sns.barplot(x=city_counts.index, y=city_counts.values, alpha=0.8)
plt.title('Number of data from each city')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('City', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame and 'City' is your column
city_counts = df['City'].value_counts()

# Create a color list
colors = ['red', 'green', 'blue']

plt.figure(figsize=(10,5))
sns.barplot(x=city_counts.index, y=city_counts.values, palette=colors, alpha=0.8)
plt.title('Number of data from each city')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('City', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```
